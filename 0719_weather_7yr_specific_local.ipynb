```python
import pandas as pd
```

```python
df_raw=pd.read_csv('weather_7yr.csv')
```

```python
#강원, 경북 지역의 기상정보만 불러오기
condition=(df_raw.지점번호==105)|(df_raw.지점번호==100)|(df_raw.지점번호==106)|(df_raw.지점번호==104)|(df_raw.지점번호==93)|(df_raw.지점번호==214)|(df_raw.지점번호==90)|(df_raw.지점번호==121)|(df_raw.지점번호==114)|(df_raw.지점번호==211)|(df_raw.지점번호==217)|(df_raw.지점번호==95)|(df_raw.지점번호==101)|(df_raw.지점번호==216)|(df_raw.지점번호==212)|(df_raw.지점번호==283)|(df_raw.지점번호==279)|(df_raw.지점번호==273)|(df_raw.지점번호==271)|(df_raw.지점번호==137)|(df_raw.지점번호==136)|(df_raw.지점번호==277)|(df_raw.지점번호==272)|(df_raw.지점번호==281)|(df_raw.지점번호==115)|(df_raw.지점번호==130)|(df_raw.지점번호==278)|(df_raw.지점번호==276)|(df_raw.지점번호==138)|(df_raw.지점번호==152)

df=df_raw[condition]

```

```python
local_name=[]
for row in df['지점번호']:
    if (row==105) | (row==100) | (row==106) | (row==104) | (row==93) | (row==214) | (row==90) | (row==121) | (row==114) | (row==211) | (row==217) | (row==95) | (row==101) | (row==216) | (row==212):
        local_name.append('강원')
    else:
        local_name.append('경북')

df.insert(0,"발생장소_시도", local_name,True)
```

```python
df_gangwon=df[df.발생장소_시도=='강원']
#강원도 내 여러 관측소의 값들의 평균값을 기상인자로 사용할 것임 
df_gangwon_mean=df_gangwon.groupby('관측일').mean()
df_gangwon_mean.reset_index(inplace=True)

#'발생장소_시도'열 만들기:어느 지역의 기상인자들인지를 표시하기 위해 '강원'으로 채워진 열을 만듬. 차후 병합을 위해 컬럼명은 산불파일과 일치시킴 
name=[]
for i in range(len(df_gangwon_mean)):
    name.append('강원')
df_gangwon_mean.insert(0,'발생장소_시도',name, True)
```

```python
df_gangwon_mean
```

```python
df_kyoungbuk=df[df.발생장소_시도=='경북']
#경북 내 여러 관측소의 값들의 평균값을 기상인자로 사용할 것임 
df_kyoungbuk_mean=df_kyoungbuk.groupby('관측일').mean()
df_kyoungbuk_mean.reset_index(inplace=True)

#'발생장소_시도'열 만들기:어느 지역의 기상인자들인지를 표시하기 위해 '경북'으로 채워진 열을 만듬. 차후 병합을 위해 컬럼명은 산불파일과 일치시킴 
name=[]
for i in range(len(df_kyoungbuk_mean)):
    name.append('경북')
df_kyoungbuk_mean.insert(0,'발생장소_시도',name, True)
```

```python
df_kyoungbuk_mean
```

```python
df=pd.concat([df_gangwon_mean,df_kyoungbuk_mean])
```

```python
df.drop(['지점번호'],axis=1, inplace=True)
```

```python
df_sort=df.sort_values('관측일')
```

```python
df_sort
```

```python
df_sort.to_csv('weather_7yr_specific_local_mean.csv',index=False)
```
