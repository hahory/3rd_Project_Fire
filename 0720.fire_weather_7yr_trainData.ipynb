```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
df1=pd.read_csv('fire_7yr_specific_local.csv')
df2=pd.read_csv('weather_7yr_specific_local_mean.csv')
#발생장소_시도,관측일로 inner join
df = pd.merge(left=df2, right=df1, how='outer', on=['발생장소_시도','관측일'])
```

```python
#산불발생여부를 구분할 열 만들기(산불발생은 1, 미발생은 0으로 표기)
fire=[]
for row in df['발생원인_구분']:
    if (row=='입') |(row=='기') | (row=='쓰') | (row=='담') | (row=='건') :
        fire.append(1)
    else:
        fire.append(0)

df.insert(0,"산불발생건수", fire,True)
```
