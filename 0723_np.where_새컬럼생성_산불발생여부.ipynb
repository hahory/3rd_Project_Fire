산불발생은 1, 산불 미발생은 0로 표기한 2014-2020년 학습용 데이터 만들기

```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv('fire_7yr_specific_local.csv')
df2=pd.read_csv('weather_7yr_specific_local_mean.csv')
#발생장소_시도,관측일로 merge
df = pd.merge(left=df2, right=df1, how='outer', on=['발생장소_시도','관측일'])
```

```python
#np.where(조건식(True/False로 반환됨),조건을 만족했을때(True일 때) 반환값,조건을 만족 안했을때(False일 때) 반환값)
#np.where(조건식)으로만 명기하는 경우 조건을 만족하는 색인만 반환
new_col=np.where(df['발생장소_관서'].isnull(),0,1)
#new_col의 값으로 '산불발생여부'라는 컬럼을 첫번째 컬럼 위치에 생성  
df.insert(0,'산불발생여부',new_col,True)
```
