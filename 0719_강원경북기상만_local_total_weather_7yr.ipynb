```python
import pandas as pd
from glob import glob
```

```python
#glob() 함수는 인자로 받은 패턴과 이름이 일치하는 모든 파일과 디렉터리의 리스트를 반환

#폴더 내에 있는 모든 csv파일 목록 불러오기
files=glob('/Users/ejju/Desktop/3rd_project/ MDIS_기상관측자료/지상_일_20220718_15746_데이터/*.csv')
total=pd.DataFrame() #빈 데이터프레임을 하나 생성하기
for file in files:
    temp=pd.read_csv(file, sep=',', encoding='euc-kr') #csv파일을 하나씩 열어 임시 데이터프레임으로 생성
    total=pd.concat([total,temp]) #전체 데이터프레임에 임시 데이터프레임들을 추가하여 넣rl
```

```python
total.to_csv('/Users/ejju/Desktop/3rd_project/ MDIS_기상관측자료/지상_일_20220718_15746_데이터/weather_7yr.csv', index=False)
```
